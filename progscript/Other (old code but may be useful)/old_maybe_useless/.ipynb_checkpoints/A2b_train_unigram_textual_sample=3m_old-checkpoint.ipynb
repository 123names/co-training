```python
import logging
import os

Dataset = "pubmed"

# read data from all file 
PATH = '../Data'
# print log
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
# parameters
cutoff = 3
```

```python
# collect cleaned tokenized data for vectorization
import com_func
from statistics import mean 

# read all files 
def readallfile(filepath):
    documents = []
    documents_length = []
    with open(filepath, 'r', encoding = 'utf8') as f:
        for line in f:
            items = line.split("\t")
            paperID = items[0]
            title = items[1].lower()
            keywords = items[2].lower()
            mesh = items[3].lower()
            abstract = items[4].lower()
            content = title+" "+keywords+" "+mesh+" "+abstract
            # remove stop words and each word must have at least 2 character
            cleaned_content = com_func.clean_line_of_raw(content)
            # collect document length information
            documents_length.append(len(cleaned_content))
            # save cleaned content to dict
            cleaned_doc = {"paperID": paperID, "cleaned_content":cleaned_content}
            # form tagged document and collect all records
            documents.append(cleaned_doc)
    f.close()
    print("Done loading files, total ", len(documents), " papers")
    return documents, documents_length

documents, documents_length = readallfile(PATH+"/"+Dataset+"/id_textual_combined.txt")
```

```python
# plot document length
print("max document length: ", max(documents_length))
print("min document length: ", min(documents_length))
print("average document length: ", mean(documents_length))
```

```python
# collect some statistic
print(len(documents))
print(documents[:2])
print(documents[0])
```

```python
# convert raw text to numerical feature vectors
# unigram are used
def dummy(doc):
    return doc
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer(min_df = cutoff,analyzer='word', tokenizer=dummy,preprocessor=dummy)
# train a dictionary to build a frquence counter
allContent = [paper["cleaned_content"] for paper in documents]
allpid = [paper["paperID"] for paper in documents]
count_matrix = count_vect.fit_transform(allContent)
print(len(count_vect.vocabulary_))
print(count_matrix.shape)
```

```python
# save the trained count vectorizer
import pickle
modelSaveDir = "../Data/"+Dataset+"/models/count/textual_sample=3m/"
if not os.path.exists(modelSaveDir):
    os.makedirs(modelSaveDir)
pickle.dump(count_vect, open(modelSaveDir+"CountVectorizer.pickle", "wb"))
pickle.dump(count_matrix, open(modelSaveDir+"trained_features.pickle", "wb"))
pickle.dump(allpid, open(modelSaveDir+"feature_pid.pickle", "wb"))
```

```python
# calculate tf-idf using count generated 
from sklearn.feature_extraction.text import TfidfVectorizer
# tf = sublinear tf, 1+log(tf) if tf >0, or 0. idf = (log(1+N)/df(t,d,C)) + 1
# df(d,t,C) means number of document d contain term t in collection C
# Each row is normalized to have unit Euclidean norm 
# reference to http://scikit-learn.org/stable/modules/feature_extraction.html#text-feature-extraction
tfidf_vectorizer = TfidfVectorizer(use_idf=True, smooth_idf=True, sublinear_tf=True, norm="l2", min_df=cutoff,
                                   analyzer='word', tokenizer=dummy,preprocessor=dummy)
tfidf_matrix = tfidf_vectorizer.fit_transform(allContent)
print(len(tfidf_vectorizer.vocabulary_))
print(tfidf_matrix.shape)
```

```python
# save the trained tf-idf
import pickle
modelSaveDir = "../Data/"+Dataset+"/models/tf-idf/textual_sample=3m/"
if not os.path.exists(modelSaveDir):
    os.makedirs(modelSaveDir)
pickle.dump(tfidf_vectorizer, open(modelSaveDir+"tf_idf_Vectorizer.pickle", "wb"))
pickle.dump(tfidf_matrix, open(modelSaveDir+"tf_idf_trained_features.pickle", "wb"))
pickle.dump(allpid, open(modelSaveDir+"feature_pid.pickle", "wb"))
```

```python
# convert tokenized text to TaggedDocument
import gensim
import multiprocessing
from gensim.models.doc2vec import TaggedDocument

formated_doc = []

for paper in documents:
    formated_doc.append(TaggedDocument(paper["cleaned_content"], [paper["paperID"]]))
    
print(formated_doc[:3])
```

```python
len(formated_doc)
```

```python
print(formated_doc[:3])
```

```python
# train the model
def train_d2v_model(document, alg=0):
    cores = multiprocessing.cpu_count()
    # train model
    # size is dimension of vector return, dm defines training algorithm, when dm = 1 is distributed memory (PV-DM),
    # pv-dbow is better than pv-md
    # when dm = 0 is distributed bag of words (PV-DBOW), 
    # sample is threshold for configuring which higher-frequency words are randomly downsampled, 
    # min_count is ignore threshold (Ignores all words with total frequency lower than min_count)
    model = gensim.models.Doc2Vec(formated_doc, dm=alg, vector_size=100, negative=5, window=5, min_count=cutoff,
                                  epochs=20, sample=1e-3, workers=cores)
    return model
    
model = train_d2v_model(documents)
```

```python
# save model
newfileDir = "../Data/"+Dataset+"/models/doc2v/textual_sample=3m/pv_dbow/"
if not os.path.exists(newfileDir):
    os.makedirs(newfileDir)
modelname = model.__str__()
print("Saving model ", modelname)
model.save(newfileDir+modelname)
print("Done")
model.delete_temporary_training_data(keep_doctags_vectors=True, keep_inference=True)
```

```python
cores = multiprocessing.cpu_count()
model_pv_dm = gensim.models.Doc2Vec(formated_doc, dm=1, vector_size=100, negative=5, window=5, min_count=cutoff,
                                  epochs=20, sample=1e-3, workers=cores)
```

```python
import unicodedata
import string
valid_filename_chars = "-_.,() %s%s" % (string.ascii_letters, string.digits)
char_limit = 255
def clean_filename(filename, whitelist=valid_filename_chars, replace=' '):
    # replace spaces
    for r in replace:
        filename = filename.replace(r,'_')
    
    # keep only valid ascii chars
    cleaned_filename = unicodedata.normalize('NFKD', filename).encode('ASCII', 'ignore').decode()
    
    # keep only whitelisted chars
    cleaned_filename = ''.join(c for c in cleaned_filename if c in whitelist)
    if len(cleaned_filename)>char_limit:
        print("Warning, filename truncated because it was over {}. Filenames may no longer be unique".format(char_limit))
    return cleaned_filename[:char_limit]
```

```python
print(model_pv_dm.__str__())
print(clean_filename(model_pv_dm.__str__()))
```

```python
# save model
newfileDir = "../Data/"+Dataset+"/models/doc2v/textual_sample=3m/pv_dm/"
if not os.path.exists(newfileDir):
    os.makedirs(newfileDir)
modelname = clean_filename(model_pv_dm.__str__())
print("Saving model ", newfileDir+modelname)
model_pv_dm.save(newfileDir+modelname)
print("Done")
# model_pv_dm.delete_temporary_training_data(keep_doctags_vectors=True, keep_inference=True)
```
