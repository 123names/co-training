```python
import warnings
warnings.filterwarnings("ignore") 

import time
import pandas as pd
import numpy as np
# suppress scientific float notation
np.set_printoptions(precision=5, suppress=True)  
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_moons
random_state = 21
cmap = 'viridis'
dot_display_size = 50
X, y = make_moons(150, noise=.07, random_state=random_state)
fig, ax = plt.subplots(figsize=(9,7))
ax.set_title('Data with ground truth labels - linear separation not possible', fontsize=18)
ax.scatter(X[:, 0], X[:, 1],c=y,s=dot_display_size, cmap=cmap)
```

```python
from scipy.cluster.hierarchy import dendrogram

def plot_dendrogram(model, **kwargs):

    # Children of hierarchical clustering
    children = model.children_

    # Distances between each pair of children
    # Since we don't have this information, we can use a uniform one for plotting
    distance = np.arange(children.shape[0])

    # The number of observations contained in each cluster level
    no_of_observations = np.arange(2, children.shape[0]+2)

    # Create linkage matrix and then plot the dendrogram
    linkage_matrix = np.column_stack([children, distance, no_of_observations]).astype(float)
    
    plt.xlabel('Samples')
    plt.ylabel('Distance (uniform one instead of real distance)')
    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)
```

```python
import scipy
from scipy.cluster import hierarchy
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist
from scipy.cluster.hierarchy import fcluster

class agglomerative_hierarchical_clustering:
    
    def __init__(self, n_clusters = 0, affinity='euclidean', linkage="single"):
        self.n_clusters = n_clusters
        self.affinity = affinity
        self.linkage = linkage
        
    def fit(self, data):
        
        if self.n_clusters > len(data):
            self.quit("Error, k is larger than the number of samples")
        print(data.shape)
        # hierarchy.linkage function combined generate pairwise distance matrix and apply linkage function 
        # 1. generate pairwise distance matrix
        # 2. apply different linkage function 
        linkage_matrix = hierarchy.linkage(data, method=self.linkage, metric=self.affinity)
        self.linkage_dist_matrix = linkage_matrix
        print(linkage_matrix.shape)
        print("First 10 in linkage matrix \n",linkage_matrix[:10])
        print("Last 5 in linkage matrix \n",linkage_matrix[-5:])
        #print(linkage_matrix)
        # extra  Cophenetic Correlation Coefficient check
        dist_matrix = pdist(data)
        print(dist_matrix.shape)
        print(dist_matrix[:10])
        c, coph_dists = cophenet(linkage_matrix, dist_matrix)
        print(c)
        print(coph_dists.shape)
        print(coph_dists[:10])
        # 3. generate labels from linkage_matrix
        if self.n_clusters >0:
            self.labels = fcluster(linkage_matrix, self.n_clusters, criterion='maxclust')
        else:
            # fcluster use inconsistent as auto distance cut-off
            # TODO: Finish auto distance cut-off
            self.labels = fcluster(linkage_matrix, criterion='inconsistent')
        
```

```python
model = agglomerative_hierarchical_clustering(n_clusters = 2)
model.fit(X)
labels = model.labels

fig, ax = plt.subplots(figsize=(9,7))
ax.set_title('scipy agglomerative Clustering', fontsize=18, fontweight='demi')
plt.scatter(X[:, 0], X[:, 1], c=labels, s=dot_display_size, cmap=cmap)
plt.show()

plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
# set cut-off to 50
max_d = 0.25  # max_d as in max_distance
plt.axhline(y=max_d, c='k') # plot a horizontal cut-off line
detail_dict = dendrogram(model.linkage_dist_matrix,
                         truncate_mode='lastp',  # show only the last p merged clusters
                         p=12,                   # show only the last p merged clusters
                         leaf_font_size=8.,      # font size for the x axis labels
                         show_contracted=True,   # to get a distribution impression in truncated branches  
)
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters=2, affinity="euclidean", linkage= "single").fit(X)
labels = model.labels_
print(model.n_components_)

fig, ax = plt.subplots(figsize=(9,7))
ax.set_title('Build in agglomerative Clustering in scikit learn', fontsize=18, fontweight='demi')
plt.scatter(X[:, 0], X[:, 1], c=labels, s=dot_display_size, cmap=cmap)
plt.show()

plt.title('Hierarchical Clustering Dendrogram')
plot_dendrogram(model, labels=model.labels_)
plt.show()
```
