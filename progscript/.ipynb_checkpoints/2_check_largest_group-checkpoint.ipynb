```python
# load the file
import sys
import io
import os

Dataset = "pubmed"

filePath = "Data/"+Dataset+"/canopies_labeled/"
listfiles = os.listdir(filePath)

LabeledRecords_original = []
labeledRecords_counter = []

for file in listfiles:
    if not file.startswith('.'):
        with open(filePath+file, 'r', encoding = 'utf8') as f:
            counter = 0
            for line in f:
                read_data = line.split("\t")
                # get ride of bad formated lines
                if(len(read_data)==13):
                    paper_detail = {"paperID": read_data[0], "authorID":read_data[1]}
                    LabeledRecords_original.append(paper_detail)
                    counter+=1
            labeledRecords_counter.append({"author_name": file, "total paper": counter})
            print("Author Name: ",file, " contains ", counter, " records.")
        f.close()
print("Total labeled records:",len(LabeledRecords_original), "records")
```

```python
from collections import Counter
sorted_labeledRecords_counter = sorted(labeledRecords_counter, key=lambda k: k['total paper']) 
print(sorted_labeledRecords_counter[-10:])
```

```python
from collections import Counter
# count number of publication based on only name
Record_NameCounter = Counter([records["author_name"] for records in LabeledRecords_original])
# print top 10 most common names that published papers in current dataset
print(Record_NameCounter.most_common(10))
```
